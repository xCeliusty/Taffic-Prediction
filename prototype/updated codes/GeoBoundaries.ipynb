```python
# Uncomment the following line to install geemap & geopandas if needed.
# !pip install geemap
#!pip install geopandas

# Reading GeoBoundaries json file
import json
from shapely.geometry import shape
import folium
import pandas as pd
# For Google CoLab
from google.colab import drive

# Opening JSON file
drive.mount('/Dataset/')
path = r"/Dataset/MyDrive/Dataset/Uber_Dataset/cairo_uberboundaries.json"
file = open(path, encoding="utf8")

# returns JSON object as
# a dictionary
data = json.load(file)

# Plotting Map
m = folium.Map([30.0444, 31.2357], zoom_start=10, tiles='cartodbpositron')
folium.GeoJson(data).add_to(m)
folium.LatLngPopup().add_to(m)

# Variable Declaration
my_cairo_centroids_x = []
my_cairo_centroids_y = []

# Iterating through the json
# list
for i in data['features']:
    my_cairo_centroids_x.append(shape(i['geometry']).centroid.x)
    my_cairo_centroids_y.append(shape(i['geometry']).centroid.y)

# Change list to dataFrame
d = {'x': my_cairo_centroids_x, 'y': my_cairo_centroids_y}
coordinates = pd.DataFrame(d)
coordinates.to_csv('data.csv')
!cp data.csv "/Dataset/MyDrive/Dataset/Uber_Dataset/regions_coordinates.csv"
m
```
